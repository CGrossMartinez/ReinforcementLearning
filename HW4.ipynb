```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt

#defining gradient decents of cost function
def grad_e(rxd_T, w_T, Rx):

    rxd_T_neg = -rxd_T 
    temp = np.matmul(w_T, Rx)
    temp2 = np.divide(temp, 2)
    temp3 = np.add(rxd_T_neg, temp2) 
    
    return temp3

#declaring and initializing weights, rsd, and reward matrices 
w_T = np.array([0, 0])
rxd_T = np.array([0.8182, 0.354])
Rx = np.array([[1, 0.8182], [0.8182, 1]])

#declaring and initializing learning rates
n1 = 0.3
n2 = 1

#declaring and initializing variables for # of iter and matrix to hold weight
#updates
num_iter = 400
w_hist1 = np.zeros((2, num_iter + 1))
w_hist1[:,0] = w_T

#loop to update the weights with lr of 0.3
for t in range(num_iter):
    w_T = w_T - n1 * grad_e(rxd_T, w_T, Rx)
    w_hist1[:,t+1] = w_T

#declaring and initializing variables for weights and matrix to hold weight
#updates
w_T = np.array([0, 0])
w_hist2 = np.zeros((2, num_iter + 1))
w_hist2[:,0] = w_T

#loop to update the weights with lr of 1
for t in range(num_iter):
    w_T = w_T - n2* grad_e(rxd_T, w_T, Rx)
    w_hist2[:,t+1] = w_T

#declaring and initializing variable for x value in plot
x = np.arange(401)

#plotting w1 value with lr of 0.3 with 400 iterations
plt.plot(x, w_hist1[0])
plt.xlabel('Iteration')
plt.ylabel('w1 value')
plt.title('Learning rate = 0.3')
plt.show()

#plotting w2 value with lr of 0.3 with 400 iterations
plt.plot(x, w_hist1[1])
plt.xlabel('Iteration')
plt.ylabel('w2 value')
plt.title('Learning rate = 0.3')
plt.show()

#plotting w1 value with lr of 1 with 400 iterations
plt.plot(x, w_hist2[0])
plt.xlabel('Iteration')
plt.ylabel('w1 value')
plt.title('Learning rate = 1')
plt.show()

#plotting w2 value with lr of 1 with 400 iterations
plt.plot(x, w_hist2[1])
plt.xlabel('Iteration')
plt.ylabel('w2 value')
plt.title('Learning rate = 1')
plt.show()
```
